```python
%load_ext autoreload
%autoreload 2
from libs.ANN import *
import numpy as np
```

```python
x1 = np.arange(1,100)
x2 = np.arange(1,100)+1
x = np.stack((x1,x2)).T

y = x1*x2/99+x1
y = np.expand_dims(y,axis=1)
```

```python
nn = NeuralNetwork([
                    Linear_layer(2, 50, bias=False),
                    Activation_layer(function='ReLU'),
    
                    Linear_layer(50, 100, bias=False),
                    Activation_layer(function='ReLU'),
    
                    Linear_layer(100, 50, bias=False),
                    Activation_layer(function='ReLU'),
    
                    Linear_layer(50, 10, bias=False),
                    Activation_layer(function='ReLU'),
    
                    Linear_layer(10, 1, bias=False),
                    ])
```

```python
param = {"lr": 1e-4, 'batch': 1, "momentum": 0.9, "mode": "train", "eps": 1e-9, "beta":(0.9, 0.999), 
         "epoch": 0, 'method': 'sgd', 't': 1, 'clip': 3.0, 'decay': 0.0}
```

```python
for _ in range(200):
    nn.train(x[:10], y[:10], param)
nn.plot_loss()
```
